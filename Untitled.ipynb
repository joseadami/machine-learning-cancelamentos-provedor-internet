### Removendo colunas que não iremos utilizar 

```python
df = df.drop(columns=['customerID','gender', 'Partner', 'Dependents', 'PhoneService','MultipleLines', 
                      'Contract','PaymentMethod', 'MonthlyCharges', 'TotalCharges', 'Contract','SeniorCitizen',
                      'PaperlessBilling'])
df
```

### Separando atribuitos para X e resultados para Y

```python
atributos = df.columns.drop('Churn')
resultados = ['Churn']



X = df[atributos]
Y = df[resultados]
```

### Dados treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25, random_state =  48)
```

```python
models = [
    ("RF", RandomForestClassifier()),
    ("DT", DecisionTreeClassifier()),
    ("LR", LogisticRegression()),
    ("Gaussian", GaussianNB()),
    ("KNC", KNeighborsClassifier()),
    ("SVC", SVC())
]
```

### Inserindo modelos dentro de um dicionário 

```python
trained_models = []
results = {'modelo': [], 'acuracia': [], 'precisao': [], 'recall': []}

for name, model in models:
    print('\n------------- ' + str(name)+ ' -------------')
    model.fit(X_train, y_train.values.ravel())
    y_pred = model.predict(X_test)
    
    #print(metrics.confusion_matrix(y_test, y_pred, labels = [1, 0]))

    trained_models.append((name, model, metrics.accuracy_score(y_test, y_pred)))
    results['modelo'].append(name)
    results['acuracia'].append(metrics.accuracy_score(y_test, y_pred))
    results['precisao'].append(metrics.precision_score(y_test, y_pred))
    results['recall'].append(metrics.recall_score(y_test, y_pred, average='macro'))
```

### Inserindo resultados dentro de uma tabela

```python
df_results = pd.DataFrame(data=results)
df_results
```
